```python
import numpy as np
import pandas as pd
import anndata as ad
```

```python
out_dir = "/Users/apple/Desktop/KB/SCSeq_LineageBarcoding/SCLineage_ConstrativeLearning/out/plot/feat_LCL_2025/cell_tag_multi"
scores_1 = np.load(f"{out_dir}/feat_celltagMulti_lambda001_unlab15_bs250_testAsPenalty/ig_cosine_centroid_global_gene_scores.npy")
scores_2 = np.load(f"{out_dir}/feat_celltagMulti_lambda005_unlab50_bs250_testAsPenalty/ig_cosine_centroid_global_gene_scores.npy")

print(scores_1.shape, scores_1.min(), scores_1.max())
print(scores_2.shape, scores_2.min(), scores_2.max())
```

```python
adata = ad.read_h5ad(h5ad_dir)  # or put the full path
genes = np.array(adata.var_names)

df = pd.DataFrame({"gene": genes, "score_001_15": scores_1, "score_005_50": scores_2})
df.head(30)
```

```python
df["rank_001_15"] = df["score_001_15"].rank(ascending=False, method="average")
df["rank_005_50"] = df["score_005_50"].rank(ascending=False, method="average")
```

```python
from scipy.stats import spearmanr

rho, p = spearmanr(df["score_001_15"], df["score_005_50"])
rho
```
