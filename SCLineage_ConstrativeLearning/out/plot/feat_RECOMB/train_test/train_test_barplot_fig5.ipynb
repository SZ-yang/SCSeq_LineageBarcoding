```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import FuncFormatter
```

## Train Test Barplot

### Biddy

```python
data = {
    'Score': [
        47.11,  # Baseline(scVI) score
        37.29,  # Supervised UMAP score
        62.71  # scCL(ours) score
    ],
    'Method': ['scVI', 'Supervised UMAP', 'LCL(ours)']
}

df = pd.DataFrame(data)

# Define custom color palette for each method
colors = {"scVI": "#DDAD3B", "Supervised UMAP": "#75A43A", "LCL(ours)": "#D43F88"}

# Create the bar plot with the data and control the width of the bars
plt.figure(figsize=(8, 6), dpi=300)
sns.barplot(x='Method', y='Score', data=df, palette=colors, width=0.5)  # Adjust bar width with 'width'
plt.grid(True, alpha=0.8, linestyle=':')

# Set y-axis label
plt.ylabel('Test Accuracy')

# Format the y-axis to display percentage
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.0f}%'))

plt.show()
```

## Compositional Analysis

```python
# Simplified DataFrame with only the first value for each method
data = {
    'Score': [
        1.3141,  # Baseline(scVI) score
        1.2521,  # Supervised UMAP score
        0.6222  # scCL(ours) score
    ],
    'Method': ['scVI', 'Supervised UMAP', 'LCL(ours)']
}

df = pd.DataFrame(data)

# Define custom color palette for each method
colors = {"scVI": "#DDAD3B", "Supervised UMAP": "#75A43A", "LCL(ours)": "#D43F88"}

# Create the bar plot with the data
plt.figure(figsize=(8, 6), dpi=300)
sns.barplot(x='Method', y='Score', data=df, palette=colors, width=0.5)

# Add grid
plt.grid(True, alpha=0.8, linestyle=':')

# Set plot title and labels
plt.xlabel('Method')
plt.ylabel('KL Divergence for Compositional Analysis')

# Show the plot
plt.show()
```
