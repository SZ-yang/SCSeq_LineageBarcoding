```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import FuncFormatter
```

## Train Test Barplot

### Biddy

```python
data = {
    'Score': [
        100-47.11,  # Baseline(scVI) score
        100-37.29,  # Supervised UMAP score
        100-62.71  # scCL(ours) score
    ],
    'Method': ['scVI', 'Supervised UMAP', 'LCL(ours)']
}

df = pd.DataFrame(data)

# Define custom color palette for each method
colors = {"scVI": "#DDAD3B", "Supervised UMAP": "#75A43A", "LCL(ours)": "#D43F88"}

# Create the bar plot with the data and control the width of the bars
plt.figure(figsize=(8, 12.6), dpi=300)
sns.barplot(x='Method', y='Score', data=df, palette=colors, width=0.7)  # Adjust bar width with 'width'
plt.grid(True, alpha=0.8, linestyle=':')
plt.gca().tick_params(axis='both', which='both', length=0)  # Hides the tick marks but keeps the grid lines
plt.gca().set_xticklabels([])  # Hides the x-axis labels
plt.gca().set_yticklabels([])  # Hides the y-axis labels
# Set y-axis label
plt.ylabel('Test Accuracy')

# Format the y-axis to display percentage
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.0f}%'))

plt.show()
```
