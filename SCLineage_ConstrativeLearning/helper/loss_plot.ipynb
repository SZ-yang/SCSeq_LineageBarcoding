```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define input and output directories
input_dir = "/Users/apple/Desktop/KB/SCSeq_LineageBarcoding2/SCSeq_LineageBarcoding/SCLineage_ConstrativeLearning/out/data/feat_414"
output_dir = "/Users/apple/Desktop/KB/SCSeq_LineageBarcoding2/SCSeq_LineageBarcoding/SCLineage_ConstrativeLearning/out/plot/feat_414"

# Load training losses
TL_bs25_temp01 = np.load(input_dir+'/feat_414__bs25_temp0.1/train_losses_bs25_tau0.1.npy')
TL_bs25_temp05 = np.load(input_dir+'/feat_414__bs25_temp0.5/train_losses_bs25_tau0.5.npy')
TL_bs25_temp10 = np.load(input_dir+'/feat_414__bs25_temp1.0/train_losses_bs25_tau1.0.npy')
TL_bs50_temp05 = np.load(input_dir+'/feat_414__bs50_temp0.5/train_losses_bs50_tau0.5.npy')
TL_bs75_temp05 = np.load(input_dir+'/feat_414__bs75_temp0.5/train_losses_bs75_tau0.5.npy')

# Load validation losses
VL_bs25_temp01 = np.load(input_dir+'/feat_414__bs25_temp0.1/val_losses_bs25_tau0.1.npy')
VL_bs25_temp05 = np.load(input_dir+'/feat_414__bs25_temp0.5/val_losses_bs25_tau0.5.npy')
VL_bs25_temp10 = np.load(input_dir+'/feat_414__bs25_temp1.0/val_losses_bs25_tau1.0.npy')
VL_bs50_temp05 = np.load(input_dir+'/feat_414__bs50_temp0.5/val_losses_bs50_tau0.5.npy')
VL_bs75_temp05 = np.load(input_dir+'/feat_414__bs75_temp0.5/val_losses_bs75_tau0.5.npy')


# Load average validation losses
avgVL_bs25_temp01 = np.load(input_dir+'/feat_414__bs25_temp0.1/avg_val_losses_bs25_tau0.1.npy')
avgVL_bs25_temp05 = np.load(input_dir+'/feat_414__bs25_temp0.5/avg_val_losses_bs25_tau0.5.npy')
avgVL_bs25_temp10 = np.load(input_dir+'/feat_414__bs25_temp1.0/avg_val_losses_bs25_tau1.0.npy')
avgVL_bs50_temp05 = np.load(input_dir+'/feat_414__bs50_temp0.5/avg_val_losses_bs50_tau0.5.npy')
avgVL_bs75_temp05 = np.load(input_dir+'/feat_414__bs75_temp0.5/avg_val_losses_bs75_tau0.5.npy')
```

```python
# Color mapping
color_map = {
    'Temp 0.1': 'blue',
    'Temp 0.5': 'green',
    'Temp 1.0': 'red'
}

# Plotting function
def plot_loss_data(loss_data, labels, title, filename):
    plt.figure(figsize=(10, 6))
    sns.set(style="whitegrid")
    epochs = range(1, len(loss_data[0]) + 1)
    for data, label in zip(loss_data, labels):
        sns.lineplot(x=epochs, y=data, label=label, color=color_map[label])  # Apply color mapping
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(title='Temperature')
    plt.savefig(f"{output_dir}/{filename}.png")
    plt.show()
    plt.close()

# Plot training losses
plot_loss_data(
    [TL_bs25_temp01, TL_bs25_temp05, TL_bs25_temp10],
    ['Temp 0.1', 'Temp 0.5', 'Temp 1.0'],
    'Training Losses',
    'training_losses'
)

# Plot validation losses
plot_loss_data(
    [VL_bs25_temp01, VL_bs25_temp05, VL_bs25_temp10],
    ['Temp 0.1', 'Temp 0.5', 'Temp 1.0'],
    'Validation Losses',
    'validation_losses'
)

# Plot average validation losses
plot_loss_data(
    [avgVL_bs25_temp01, avgVL_bs25_temp05, avgVL_bs25_temp10],
    ['Temp 0.1', 'Temp 0.5', 'Temp 1.0'],
    'Average Validation Losses',
    'avg_validation_losses'
)
```
